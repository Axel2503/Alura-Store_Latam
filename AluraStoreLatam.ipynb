# 2. Ventas por categor칤a

```python
import pandas as pd

file_map = {
    "alura-es-cursos/challenge1-data-science-latam/challenge1-data-science-latam-c280e292765416420f1bffa6f324a07ef021e1b0/base-de-datos-challenge1-latam/tienda_1 .csv": "Tienda 1",
    "alura-es-cursos/challenge1-data-science-latam/challenge1-data-science-latam-c280e292765416420f1bffa6f324a07ef021e1b0/base-de-datos-challenge1-latam/tienda_2.csv": "Tienda 2",
    "alura-es-cursos/challenge1-data-science-latam/challenge1-data-science-latam-c280e292765416420f1bffa6f324a07ef021e1b0/base-de-datos-challenge1-latam/tienda_3.csv": "Tienda 3",
    "alura-es-cursos/challenge1-data-science-latam/challenge1-data-science-latam-c280e292765416420f1bffa6f324a07ef021e1b0/base-de-datos-challenge1-latam/tienda_4.csv": "Tienda 4"
}

all_data = []

# Iteramos sobre el diccionario
for filepath, store_name in file_map.items():
    try:
        df_temp = pd.read_csv(filepath)
        # 춰Importante! A침adimos la columna que identifica la tienda
        df_temp['tienda'] = store_name
        all_data.append(df_temp)
    except FileNotFoundError:
        print(f"Advertencia: No se encontr칩 el archivo {filepath}")
    except Exception as e:
        print(f"Error cargando {filepath}: {e}")

# Combinamos todos los DataFrames en uno solo
df_ventas = pd.concat(all_data, ignore_index=True)

print("Datos cargados y combinados:")
print(df_ventas.head())
print(f"\nTiendas en el DataFrame: {df_ventas['tienda'].unique()}")

columna_categoria = 'categoria'

# 1. Agrupamos por 'tienda' y por la columna de categor칤a
# 2. Usamos .size() para contar el n칰mero de filas (ventas) en cada grupo
# 3. reset_index() convierte el resultado en un DataFrame limpio
try:
    ventas_por_categoria = df_ventas.groupby(['tienda', columna_categoria]).size().reset_index(name='total_ventas')

    print("\n--- Total de Ventas por Tienda y Categor칤a ---")
    print(ventas_por_categoria)

except KeyError:
    print(f"\nError: La columna '{columna_categoria}' no existe en los datos.")
    print(f"Columnas disponibles: {df_ventas.columns.tolist()}")

# Ordenamos los valores, primero por tienda (alfab칠tico)
# y luego por 'total_ventas' (de mayor a menor)
populares_ordenadas = ventas_por_categoria.sort_values(
    by=['tienda', 'total_ventas'],
    ascending=[True, False]
)

print("\n--- Categor칤as M치s Populares (ordenadas) ---")
print(populares_ordenadas)

top_3_por_tienda = populares_ordenadas.groupby('tienda').head(3)

print("\n--- 游끥 TOP 3 Categor칤as por Tienda 游끥 ---")
print(top_3_por_tienda)
```

# 3. Calificaci칩n promedio de la tienda


```python
columna_calificacion = 'Calificacion'

try:
    # 1. Agrupamos solo por 'tienda'
    # 2. Seleccionamos la columna de calificaci칩n
    # 3. Calculamos la media (promedio)
    calificacion_promedio_tienda = df_ventas.groupby('tienda')[columna_calificacion].mean()

    # Redondeamos a 2 decimales para una mejor lectura
    calificacion_promedio_tienda = calificacion_promedio_tienda.round(2)

    # Lo convertimos a DataFrame para una impresi칩n m치s clara
    calificacion_promedio_tienda = calificacion_promedio_tienda.reset_index(name='calificacion_promedio')

    # Ordenamos de mejor a peor calificaci칩n
    calificacion_promedio_tienda = calificacion_promedio_tienda.sort_values(
        by='calificacion_promedio',
        ascending=False
    )

    print("\n--- 救넖잺 Calificaci칩n Promedio por Tienda ---")
    print(calificacion_promedio_tienda)

except KeyError:
    print(f"\nError: La columna '{columna_calificacion}' no existe en los datos.")
    print("Por favor, revisa el nombre de la columna de calificaciones.")
    print(f"Columnas disponibles: {df_ventas.columns.tolist()}")
except TypeError:
    print(f"\nError: La columna '{columna_calificacion}' parece no ser num칠rica.")
    print("Aseg칰rate de que esa columna contenga n칰meros para poder calcular el promedio.")
```

# 4. Productos m치s y menos vendidos

```python
columna_producto = 'Producto'

try:
    # 1. Agrupar por tienda y producto, y contar las ventas de cada uno
    ventas_por_producto = df_ventas.groupby(['tienda', columna_producto]).size().reset_index(name='total_ventas')

    # 2. Ordenar por tienda y luego por total_ventas (descendente)
    ventas_ordenadas_desc = ventas_por_producto.sort_values(
        by=['tienda', 'total_ventas'],
        ascending=[True, False]
    )

    # 3. Ordenar por tienda y luego por total_ventas (ascendente)
    ventas_ordenadas_asc = ventas_por_producto.sort_values(
        by=['tienda', 'total_ventas'],
        ascending=[True, True]
    )

    # 4. Obtener el Top 5 de CADA tienda
    top_5_por_tienda = ventas_ordenadas_desc.groupby('tienda').head(5)

    # 5. Obtener el Bottom 5 de CADA tienda
    bottom_5_por_tienda = ventas_ordenadas_asc.groupby('tienda').head(5)


    print("\n--- 游늳 Top 5 Productos M츼S Vendidos por Tienda ---")
    print(top_5_por_tienda.to_string()) # .to_string() para ver todas las filas

    print("\n\n--- 游늴 5 Productos MENOS Vendidos por Tienda ---")
    print(bottom_5_por_tienda.to_string())


except KeyError:
    print(f"\nError: La columna '{columna_producto}' no existe en los datos.")
    print("Por favor, revisa el nombre de la columna de productos.")
    print(f"Columnas disponibles: {df_ventas.columns.tolist()}")

except Exception as e:
    print(f"Ocurri칩 un error inesperado: {e}")

import matplotlib.pyplot as plt
import os

N_TOP_PRODUCTOS = 10
GRAFICOS_DIR = "graficos_top_productos"

try:
    os.makedirs(GRAFICOS_DIR, exist_ok=True)

    tiendas = df_ventas['tienda'].unique()

    print(f"\n--- 游늵 Generando {len(tiendas)} gr치ficos en la carpeta '{GRAFICOS_DIR}/' ---")

    for tienda in tiendas:
        # 1. Filtrar los datos para esta tienda y tomar el Top N
        top_n_tienda = ventas_ordenadas_desc[
            ventas_ordenadas_desc['tienda'] == tienda
        ].head(N_TOP_PRODUCTOS)

        # 2. Invertir el orden (para que plt.barh() muestre el #1 arriba)
        top_n_tienda = top_n_tienda.iloc[::-1]

        # 3. Crear la figura
        plt.figure(figsize=(10, 8)) # Tama침o (ancho, alto)

        # 4. Crear el gr치fico de barras horizontal
        plt.barh(
            top_n_tienda[columna_producto],
            top_n_tienda['total_ventas'],
            color='teal'
        )

        # 5. A침adir t칤tulo y etiquetas
        plt.title(f"Top {N_TOP_PRODUCTOS} Productos M치s Vendidos - {tienda}", fontsize=16)
        plt.xlabel("Total de Unidades Vendidas", fontsize=12)
        plt.ylabel("Producto", fontsize=12)

        # 6. Asegurar que todo quepa bien
        plt.tight_layout()

        # 7. Guardar el gr치fico en un archivo
        filename = os.path.join(GRAFICOS_DIR, f"top_{N_TOP_PRODUCTOS}_productos_{tienda.replace(' ', '_')}.png")
        plt.savefig(filename)

        # 8. Cerrar la figura para liberar memoria
        plt.close()

    print("춰Gr치ficos generados con 칠xito!")

except NameError:
    print("\nError: Aseg칰rate de haber ejecutado el bloque de 'An치lisis' (paso 1) primero.")
except Exception as e:
    print(f"\nOcurri칩 un error durante la generaci칩n de gr치ficos: {e}")
```

# 5. Env칤o promedio por tienda

```python
# Asumimos que 'df_ventas' ya existe y tiene los datos de todas las tiendas.

columna_costo_envio = 'costo_envio'

try:
    # 1. Agrupamos solo por 'tienda'
    # 2. Seleccionamos la columna de costo de env칤o
    # 3. Calculamos la media (promedio)
    costo_envio_promedio = df_ventas.groupby('tienda')[columna_costo_envio].mean()

    # Redondeamos a 2 decimales para una mejor lectura
    costo_envio_promedio = costo_envio_promedio.round(2)

    # Lo convertimos a DataFrame para una impresi칩n m치s clara
    costo_envio_promedio = costo_envio_promedio.reset_index(name='costo_envio_promedio')

    # Ordenamos de mayor a menor costo promedio
    costo_envio_promedio = costo_envio_promedio.sort_values(
        by='costo_envio_promedio',
        ascending=False
    )

    print("\n--- 游뚵 Costo de Env칤o Promedio pagado por el Cliente (por Tienda) ---")
    print(costo_envio_promedio)

except KeyError:
    print(f"\nError: La columna '{columna_costo_envio}' no existe en los datos.")
    print("Por favor, revisa el nombre de la columna de costos de env칤o.")
    print(f"Columnas disponibles: {df_ventas.columns.tolist()}")
except TypeError:
    print(f"\nError: La columna '{columna_costo_envio}' parece no ser num칠rica.")
    print("Aseg칰rate de que esa columna contenga n칰meros para poder calcular el promedio.")
```
